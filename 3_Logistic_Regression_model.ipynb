```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    # 直接返回sigmoid函数
    return 1. / (1. + np.exp(-x))
 
 
def plot_sigmoid():
    # param:起点，终点，间距
    x = np.arange(-8, 8, 0.2)
    y = sigmoid(x)
    plt.plot(x, y)
    plt.show()
 
 
if __name__ == '__main__':
    plot_sigmoid()
```

对于上面的图对于任意的x值都可以被映射到从0到1之间，可以吧输出值当做一个概率值    
## 什么是逻辑回归  
sigmoid函数是一个概率分布函数，给定某个输入，他将输出为一个概率值  
## 逻辑回归损失函数  
平方差所惩罚的食欲损失为同一数量的情形  
对于分类问题，使用交叉熵损失函数会更有效，交叉熵会输出一个更大的“损失”  
## 交叉熵损失函数
交叉熵刻画的是实际输出（概率）与期望输出（概率）的距离，也就是交叉熵的值越小，两个概率分布就越接近。假设概率分布p为实际输出，H（p,q）为交叉熵，则：  
$$ H(p,q) = -\sum_{x} p(x)log q(x)$$  
pytorch里面，我们使用 nn.BCELoss()来计算二元交叉熵损失  


```python
data = pd.read_csv('dataset/credit-a.csv',header=None)
```

```python
X = data.iloc[:,:-1]#pandas的方法，按位置取值
```

```python
X
```

```python
Y = data.iloc[:,-1].replace(-1,0)#将-1替换为0
```

```python
Y
```

```python
Y.unique()
```

```python
X = torch.from_numpy(X.values).type(torch.float32)
```

```python
X.shape
```

```python
Y= torch.from_numpy(Y.values.reshape(-1,1)).type(torch.float32)
```

```python
Y.size()
```

```python
from torch import nn
```

```python
model = nn.Sequential(
        nn.Linear(15,1),   #输入15列，特征为15个，输出为y,注意不层构建时中间要有“,”
        nn.Sigmoid()
)
```

```python
model
```

```python
loss_fn = nn.BCELoss()
```

```python
opt = torch.optim.Adam(model.parameters(),lr = 0.001)
```

```python
batches = 16
no_of_batch = 653//16 #整除
epoches = 1000
```

```python
for epoch in range(epoches):
    for i in range(no_of_batch):
        start = i*batches
        end = start + batches
        x = X[start:end]
        y = Y[start:end]
        y_pred = model(x)
        loss = loss_fn(y_pred,y)
        opt.zero_grad()
        loss.backward()
        opt.step()
```

```python
model.state_dict()#模型保存  x1*w1+x2*w2+`....+b
```
